```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# ==============================================================================
# --- EJERCICIO 1: Distribución Exponencial (lambda = 1.5) ---
# ==============================================================================

lambda_param = 1.5
n_muestra = 1000
n_bootstrap = 10000

print(f"--- Ejercicio 1: Distribución Exponencial (lambda = {lambda_param}) ---")

# --- Generar la muestra y estimar parámetros (a) ---
np.random.seed(42)
# scale = 1/lambda
muestra = np.random.exponential(scale=1/lambda_param, size=n_muestra)

media_est = np.mean(muestra)
mediana_est = np.median(muestra)
varianza_est = np.var(muestra, ddof=1)

print("\na) Estimaciones a partir de la muestra:")
print(f"  Media Muestral: {media_est:.4f}")
print(f"  Mediana Muestral: {mediana_est:.4f}")
print(f"  Varianza Muestral: {varianza_est:.4f}")

# --- Estimación por Bootstrap (b) ---
medias_bootstrap = np.empty(n_bootstrap)
medianas_bootstrap = np.empty(n_bootstrap)
varianzas_bootstrap = np.empty(n_bootstrap)

for i in range(n_bootstrap):
    re_muestra = np.random.choice(muestra, size=n_muestra, replace=True)
    medias_bootstrap[i] = np.mean(re_muestra)
    medianas_bootstrap[i] = np.median(re_muestra)
    varianzas_bootstrap[i] = np.var(re_muestra, ddof=1)

var_media_boot = np.var(medias_bootstrap)
bias_media_boot = np.mean(medias_bootstrap) - media_est
mse_media_boot = var_media_boot + bias_media_boot**2

var_mediana_boot = np.var(medianas_bootstrap)
bias_mediana_boot = np.mean(medianas_bootstrap) - mediana_est
mse_mediana_boot = var_mediana_boot + bias_mediana_boot**2

var_varianza_boot = np.var(varianzas_bootstrap)
bias_varianza_boot = np.mean(varianzas_bootstrap) - varianza_est
mse_varianza_boot = var_varianza_boot + bias_varianza_boot**2

print("\nb) Estimación por Bootstrap de la Varianza y MSE del estimador:")
print("  | Estimador | Varianza (Bootstrap) | Sesgo (Bootstrap) | MSE (Bootstrap) |")
print("  |-----------|----------------------|-------------------|-----------------|")
print(f"  | Media     | {var_media_boot:.6f}             | {bias_media_boot:.6f}          | {mse_media_boot:.6f}        |")
print(f"  | Mediana   | {var_mediana_boot:.6f}             | {bias_mediana_boot:.6f}          | {mse_mediana_boot:.6f}        |")
print(f"  | Varianza  | {var_varianza_boot:.6f}             | {bias_varianza_boot:.6f}          | {mse_varianza_boot:.6f}        |")

# ==============================================================================
# --- EJERCICIO 2: Costo Total del Proyecto (CT = C1 + C2) ---
# ==============================================================================

# Parámetros del ejercicio 2 (Redefinidos para asegurar la ejecución)
mu_C1 = 100000
sigma_C1 = 500
a = 800000
lambda_T = 1/5
n_muestra_CT = 1000 # Tamaño de la muestra (inciso a)
n_bootstrap_CT = 10000 # Número de re-muestreos para el inciso c)

print(f"\n{'='*60}")
print(f"--- Ejercicio 2: Costo Total del Proyecto (CT = C1 + C2) ---")
print(f"{'='*60}")

# --- Paso 1: a) Generar la Muestra para CT ---
np.random.seed(42)

T_muestra = np.random.exponential(scale=1/lambda_T, size=n_muestra_CT)
C2_muestra = a * T_muestra
C1_muestra = np.random.normal(loc=mu_C1, scale=sigma_C1, size=n_muestra_CT)
CT_muestra = C1_muestra + C2_muestra

print(f"\na) Generada muestra de {n_muestra_CT} para el Costo Total (CT).")


# --- Paso 2: b) Estimar la media y la desviación estándar para CT ---
media_CT_est = np.mean(CT_muestra)
std_CT_est = np.std(CT_muestra, ddof=1)
varianza_CT_est = std_CT_est**2

print("\nb) Estimaciones a partir de la muestra:")
print(f"  Media Muestral (Estimación de la Media de CT): {media_CT_est:,.2f}")
print(f"  Desviación Estándar Muestral (Est. de la Desv. Std.): {std_CT_est:,.2f}")

# --- Paso 3: c) Estimar Varianza y MSE de los estimadores usando Bootstrap ---
medias_bootstrap_CT = np.empty(n_bootstrap_CT)
varianzas_bootstrap_CT = np.empty(n_bootstrap_CT)

for i in range(n_bootstrap_CT):
    re_muestra = np.random.choice(CT_muestra, size=n_muestra_CT, replace=True)
    medias_bootstrap_CT[i] = np.mean(re_muestra)
    varianzas_bootstrap_CT[i] = np.var(re_muestra, ddof=1)

# Cálculo de la Varianza del Estimador
var_media_boot_CT = np.var(medias_bootstrap_CT)
var_varianza_boot_CT = np.var(varianzas_bootstrap_CT)

# Cálculo del Sesgo (Bias)
bias_media_boot_CT = np.mean(medias_bootstrap_CT) - media_CT_est
bias_varianza_boot_CT = np.mean(varianzas_bootstrap_CT) - varianza_CT_est

# Cálculo del Error Cuadrático Medio (MSE)
mse_media_boot_CT = var_media_boot_CT + bias_media_boot_CT**2
mse_varianza_boot_CT = var_varianza_boot_CT + bias_varianza_boot_CT**2

print("\nc) Estimación por Bootstrap de la Varianza y MSE del estimador:")
print("  | Estimador | Varianza (Bootstrap) | Sesgo (Bootstrap) | MSE (Bootstrap) |")
print("  |-----------|----------------------|-------------------|-----------------|")
print(f"  | Media     | {var_media_boot_CT:,.2f}             | {bias_media_boot_CT:,.2f}          | {mse_media_boot_CT:,.2f}        |")
print(f"  | Varianza  | {var_varianza_boot_CT:,.2f}             | {bias_varianza_boot_CT:,.2f}          | {mse_varianza_boot_CT:,.2f}        |")


# ==============================================================================
# --- EJERCICIO 3: Impacto del Error en los Parámetros ---
# ==============================================================================

print(f"\n{'='*60}")
print(f"--- Ejercicio 3: Impacto del Error en los Parámetros ---")
print(f"{'='*60}")

# Valor esperado del costo total base (Analítico)
E_CT_base = mu_C1 + a * (1/lambda_T)
error = 0.20 # Error del 20%

print(f"Valor Esperado (Base): E[CT] = {E_CT_base:,.2f}")

# --- Paso 1: a) Error del 20% en el parámetro 'a' ---
a_nuevo = a * (1 + error)
E_CT_a_error = mu_C1 + a_nuevo * (1/lambda_T)
cambio_porcentual_a = (E_CT_a_error - E_CT_base) / E_CT_base

print("\na) Error del 20% en el parámetro 'a':")
print(f"  Nuevo Valor Esperado: E[CT]' = {E_CT_a_error:,.2f}")
print(f"  El cambio porcentual en E[CT] es: {cambio_porcentual_a * 100:.4f}%")

# --- Paso 2: b) Error del 20% en el parámetro sigma de C1 ---
# El valor esperado de C1 (mu_C1) es independiente de sigma_C1.
E_CT_sigma_error = E_CT_base
cambio_porcentual_sigma = (E_CT_sigma_error - E_CT_base) / E_CT_base

print("\nb) Error del 20% en el parámetro sigma de C1:")
print(f"  Nuevo Valor Esperado: E[CT]' = {E_CT_sigma_error:,.2f}")
print(f"  El cambio porcentual en E[CT] es: {cambio_porcentual_sigma * 100:.4f}%")
```
