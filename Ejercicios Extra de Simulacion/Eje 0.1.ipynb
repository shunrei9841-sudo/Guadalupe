```python
from random import random
from math import exp

def poisson(lam):
    u = random()
    pk = exp(-lam)
    s = pk
    k = 1
    while s < u:
        pk = (lam / k) * pk
        s = s + pk  # CAMBIO CLAVE: Antes tenÃ­as '*' ahora es '+'
        k = k + 1
    return k - 1

lam = 4
for i in range(100):
    print(poisson(lam), ' ', end='')
```
