```python
from random import random
from math import exp
def poisson(lam):
    u=random()# Genera un número aleatorio entre 0 y 1
    pk=exp(-lam)# Calcula P(X = 0) usando la fórmula e^(-λ)
    s=pk# 's' es la probabilidad acumulada (empezamos con P(X=0))
    k=1# Contador para probar el siguiente valor (X = 1)
    while s<u:
      pk=(lam/k)*pk
      s=s+pk
      k=k+1
    return k-1
lam=4
for i in range(100):
  print(poisson(lam),' ',end='')
```
