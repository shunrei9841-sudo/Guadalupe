```python
import numpy as np

# Número de simulaciones
N = 1000000

# -----------------------------
# A) Monte Carlo básico
# -----------------------------
U = np.random.uniform(0, 2, N)
f_U = np.exp(-U**2)

I_basic = 2 * np.mean(f_U)
Var_basic = 4 * np.var(f_U) / N  # var del estimador

print("A) Estimación Monte Carlo básica =", I_basic)
print("Varianza del estimador =", Var_basic)


# -----------------------------
# B) Variables antitéticas
# -----------------------------
U = np.random.uniform(0, 2, N//2)
U_ant = 2 - U

f1 = np.exp(-U**2)
f2 = np.exp(-U_ant**2)

f_avg = 0.5 * (f1 + f2)

I_ant = 2 * np.mean(f_avg)
Var_ant = 4 * np.var(f_avg) / (N//2)

reduction_ant = 100 * (1 - Var_ant / Var_basic)

print("\nB) Con variables antitéticas =", I_ant)
print("Varianza antitética =", Var_ant)
print("Reducción de varianza = %.2f %% " % reduction_ant)


# -----------------------------
# C) Variable de control: X²
# -----------------------------
U = np.random.uniform(0, 2, N)
f_U = np.exp(-U**2)
control = U**2

# Valor esperado exacto de X^2 en [0,2]
E_control = np.mean(control)  # podría usarse 4/3 pero usamos estimación exacta MC

# Cálculo del coeficiente óptimo beta*
cov = np.cov(f_U, control, bias=True)[0,1]
var_control = np.var(control)
beta = cov / var_control

# Estimador de control
f_controlled = f_U - beta * (control - E_control)

I_control = 2 * np.mean(f_controlled)
Var_control = 4 * np.var(f_controlled) / N

reduction_control = 100 * (1 - Var_control / Var_basic)

print("\nC) Con variable de control =", I_control)
print("Varianza control =", Var_control)
print("Reducción de varianza = %.2f %% " % reduction_control)
```
