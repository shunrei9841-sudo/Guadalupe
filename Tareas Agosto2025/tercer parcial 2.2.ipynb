```python
import numpy as np

# Parámetros del ejercicio 2
mu_C1 = 100000      # Media de C1 (Normal)
sigma_C1 = 500      # Desviación estándar de C1 (Normal)
a = 800000          # Factor de C2
lambda_T = 1/5      # Parámetro lambda de la Exponencial para T

# Parámetros de la simulación
n_muestra = 1000    # Tamaño de la muestra (inciso a)
n_bootstrap = 10000 # Número de re-muestreos (inciso c)

print(f"--- Ejercicio 2: Solución del Costo Total (CT = C1 + C2) ---")
print(f"Parámetros: C1 ~ N(100k, 500); C2 = 800k * T, T ~ Exp(1/5)")

# --- a) Generar una muestra de tamaño 1,000 para el costo total (CT) ---
np.random.seed(42)

# Simulación de C2
T_muestra = np.random.exponential(scale=1/lambda_T, size=n_muestra)
C2_muestra = a * T_muestra
# Simulación de C1
C1_muestra = np.random.normal(loc=mu_C1, scale=sigma_C1, size=n_muestra)
# Costo Total
CT_muestra = C1_muestra + C2_muestra

print(f"\na) Generada muestra de tamaño {n_muestra} para el Costo Total.")


# --- b) Estimar la media y la desviación estándar para CT ---
media_CT_est = np.mean(CT_muestra)
std_CT_est = np.std(CT_muestra, ddof=1)
varianza_CT_est = std_CT_est**2

print("\nb) Estimaciones a partir de la muestra (inciso a):")
print(f"  Media Muestral (Est. de E[CT]): {media_CT_est:,.2f}")
print(f"  Desviación Estándar Muestral (Est. de la Desv. Std.): {std_CT_est:,.2f}")


# --- c) Usar simulación de Monte Carlo y "bootstrap" para estimar la varianza y el error cuadrático medio (MSE) ---
medias_bootstrap = np.empty(n_bootstrap)
varianzas_bootstrap = np.empty(n_bootstrap)

# Simulación Bootstrap (Monte Carlo)
for i in range(n_bootstrap):
    # La indentación dentro del loop 'for' debe ser EXACTAMENTE la misma.
    re_muestra = np.random.choice(CT_muestra, size=n_muestra, replace=True)
    medias_bootstrap[i] = np.mean(re_muestra)
    varianzas_bootstrap[i] = np.var(re_muestra, ddof=1)

# Cálculo de la Varianza del Estimador (Bootstrap)
var_media_boot = np.var(medias_bootstrap)
var_varianza_boot = np.var(varianzas_bootstrap)

# Cálculo del Sesgo (Bias) del Estimador
bias_media_boot = np.mean(medias_bootstrap) - media_CT_est
bias_varianza_boot = np.mean(varianzas_bootstrap) - varianza_CT_est

# Cálculo del Error Cuadrático Medio (MSE)
mse_media_boot = var_media_boot + bias_media_boot**2
mse_varianza_boot = var_varianza_boot + bias_varianza_boot**2

print("\nc) Estimación por Bootstrap de la Varianza y MSE del estimador:")
print("  | Estimador | Varianza (Bootstrap) | Sesgo (Bootstrap) | MSE (Bootstrap) |")
print("  |-----------|----------------------|-------------------|-----------------|")
print(f"  | Media     | {var_media_boot:,.2f}             | {bias_media_boot:,.2f}          | {mse_media_boot:,.2f}        |")
print(f"  | Varianza  | {var_varianza_boot:,.2f}             | {bias_varianza_boot:,.2f}          | {mse_varianza_boot:,.2f}        |")
```
