```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parámetros del ejercicio
lambda_param = 1.5
n_muestra = 1000
n_bootstrap = 10000  # Número de re-muestreos para bootstrap

print(f"--- Ejercicio 1: Distribución Exponencial (lambda = {lambda_param}) ---")

# --- Paso 1: Generar la muestra ---
# La media teórica es 1/lambda = 1/1.5 ≈ 0.6667
# La varianza teórica es 1/(lambda^2) = 1/(1.5^2) ≈ 0.4444
# La mediana teórica es ln(2)/lambda ≈ 0.4621
np.random.seed(42) # Para reproducibilidad
muestra = np.random.exponential(scale=1/lambda_param, size=n_muestra)

print(f"\nGenerada muestra de {n_muestra} números con distribución Exponencial.")
#
# --- Paso 2: a) Estimar la media, mediana y varianza a partir de la muestra ---
# Estimadores de Máxima Verosimilitud (MLE) o Estimadores de Momento
media_est = np.mean(muestra)
mediana_est = np.median(muestra)
varianza_est = np.var(muestra, ddof=1) # ddof=1 para la varianza muestral insesgada

print("\na) Estimaciones a partir de la muestra:")
print(f"  Media Muestral (Est. de la media): {media_est:.4f}")
print(f"  Mediana Muestral (Est. de la mediana): {mediana_est:.4f}")
print(f"  Varianza Muestral (Est. de la varianza): {varianza_est:.4f}")

# --- Paso 3: b) Estimar por simulación de Monte Carlo (Bootstrap) ---
# Usamos el bootstrap para estimar la varianza y el error cuadrático medio (MSE)
# de los estimadores obtenidos en el paso (a).

# Almacenar las estimaciones bootstrap
medias_bootstrap = np.empty(n_bootstrap)
medianas_bootstrap = np.empty(n_bootstrap)
varianzas_bootstrap = np.empty(n_bootstrap)

for i in range(n_bootstrap):
    # Generar una re-muestra con reemplazo del tamaño de la muestra original
    re_muestra = np.random.choice(muestra, size=n_muestra, replace=True)

    # Calcular y almacenar las estadísticas de la re-muestra
    medias_bootstrap[i] = np.mean(re_muestra)
    medianas_bootstrap[i] = np.median(re_muestra)
    varianzas_bootstrap[i] = np.var(re_muestra, ddof=1)

# Cálculo de la Varianza del Estimador (Bootstrap)
var_media_boot = np.var(medias_bootstrap)
var_mediana_boot = np.var(medianas_bootstrap)
var_varianza_boot = np.var(varianzas_bootstrap)

# Cálculo del Sesgo (Bias) del Estimador (Bootstrap)
# Sesgo = E[Estimador] - Parámetro_Estimado_en_a
bias_media_boot = np.mean(medias_bootstrap) - media_est
bias_mediana_boot = np.mean(medianas_bootstrap) - mediana_est
bias_varianza_boot = np.mean(varianzas_bootstrap) - varianza_est

# Cálculo del Error Cuadrático Medio (MSE) del Estimador (Bootstrap)
# MSE = Varianza + Sesgo^2
mse_media_boot = var_media_boot + bias_media_boot**2
mse_mediana_boot = var_mediana_boot + bias_mediana_boot**2
mse_varianza_boot = var_varianza_boot + bias_varianza_boot**2

print("\nb) Estimación por Bootstrap de la Varianza y MSE del estimador:")
print("  | Estimador | Varianza (Bootstrap) | Sesgo (Bootstrap) | MSE (Bootstrap) |")
print("  |-----------|----------------------|-------------------|-----------------|")
print(f"  | Media     | {var_media_boot:.6f}             | {bias_media_boot:.6f}          | {mse_media_boot:.6f}        |")
print(f"  | Mediana   | {var_mediana_boot:.6f}             | {bias_mediana_boot:.6f}          | {mse_mediana_boot:.6f}        |")
print(f"  | Varianza  | {var_varianza_boot:.6f}             | {bias_varianza_boot:.6f}          | {mse_varianza_boot:.6f}        |")
```
