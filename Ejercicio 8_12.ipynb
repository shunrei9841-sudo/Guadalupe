```python
import numpy as np
from sympy import symbols, simplify
from fractions import Fraction

#  1. Definición de Coeficientes
# Los coeficientes obtenidos de la resolución del sistema lineal:
# alpha_0 = 0 (de u(0)=0)
# alpha_1 = 2 (de u'(0)=2)
# alpha_2 y alpha_3 provienen de resolver 18*a2 + 25*a3 = -4 y 3*a2 + 7*a3 = -1
alpha_0 = 0
alpha_1 = 2
alpha_2 = Fraction(-1, 17)
alpha_3 = Fraction(-2, 17)

print("Coeficientes Encontrados")
print(f"alpha_0 = {alpha_0}")
print(f"alpha_1 = {alpha_1}")
print(f"alpha_2 = {alpha_2}")
print(f"alpha_3 = {alpha_3}")

# 2. Definición de la Función con SymPy
# La solución aproximada es u_3(x) = alpha_0 + alpha_1*x + alpha_2*x^2 + alpha_3*x^3
x = symbols('x')
u3_x = alpha_0 + alpha_1 * x + alpha_2 * x**2 + alpha_3 * x**3

# Simplificamos la expresión para obtener la forma final
u3_simplified = simplify(u3_x)

# 3. Verificación de Colocación
print("\nVerificación en Puntos de Colocación")
# El residual epsilon(x) debe ser cero en los puntos x=1/2 y x=1.
# Aquí evaluamos el polinomio en los puntos.
print(f"u_3(1/2) = {u3_simplified.subs(x, Fraction(1, 2))}")
print(f"u_3(1) = {u3_simplified.subs(x, 1)}")

# 4. Impresión de la Solución Final
print("\nSolución Aproximada Final")
print("La solución aproximada (polinomio cúbico) es:")
print(f"u_3(x) = {u3_simplified}")
```
