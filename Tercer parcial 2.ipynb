```python
import numpy as np # <-- ¡AGREGAR ESTA LÍNEA!

# --- Ejercicio 2: Simulación del Costo Total ---
n_simulaciones = 100000 # Un gran número de simulaciones Monte Carlo
np.random.seed(42)

# Parámetros de C1
mu_C1 = 100000
sigma_C1 = 500

# Parámetros de C2
a = 800000
lambda_T = 1/5
# La media de T es 1/lambda_T = 5

print(f"\n--- Ejercicio 2: Simulación del Costo Total (Monte Carlo) ---")

# --- Paso 1: Simular C1 (Normal) ---
C1_sim = np.random.normal(loc=mu_C1, scale=sigma_C1, size=n_simulaciones)

# --- Paso 2: Simular T (Exponencial) y calcular C2 ---
# np.random.exponential usa 'scale' que es 1/lambda
T_sim = np.random.exponential(scale=1/lambda_T, size=n_simulaciones)
C2_sim = a * T_sim


# --- Paso 3: Calcular el Costo Total (CT) ---
CT_sim = C1_sim + C2_sim

# --- Paso 4: a) Estimar el Costo Total Promedio y Varianza ---
media_CT_est = np.mean(CT_sim)
varianza_CT_est = np.var(CT_sim, ddof=1)

print("\na) Estimación del Costo Total a partir de la simulación:")
print(f"  Costo Total Promedio (Media de CT): {media_CT_est:,.2f}")
print(f"  Varianza del Costo Total (Var de CT): {varianza_CT_est:,.2f}")

# --- Paso 5: b) Comparación con los valores teóricos (Analíticos) ---
# Usando las propiedades de la esperanza:
media_C2_teo = a * (1/lambda_T)
media_CT_teo = mu_C1 + media_C2_teo

# Usando las propiedades de la varianza: (C1 y C2 son independientes)
varianza_CT_teo = sigma_C1**2 + (a**2) * (1/lambda_T)**2

print("\nb) Comparación con los valores teóricos (Analíticos):")
print(f"  Media Teórica de CT: {media_CT_teo:,.2f}")
print(f"  Varianza Teórica de CT: {varianza_CT_teo:,.2f}")

# --- Paso 6: c) Usar el teorema del límite central para estimar el error medio ---
desviacion_CT_sim = np.std(CT_sim)
error_estandar = desviacion_CT_sim / np.sqrt(n_simulaciones)

IC_95_inferior = media_CT_est - 1.96 * error_estandar
IC_95_superior = media_CT_est + 1.96 * error_estandar

print("\nc) Estimación del error para la media (Teorema del Límite Central):")
print(f"  Desviación Estándar de la muestra CT: {desviacion_CT_sim:,.2f}")
print(f"  Error Estándar (SE): {error_estandar:,.2f}")
print(f"  Intervalo de Confianza del 95% para la Media:")
print(f"    [{IC_95_inferior:,.2f}, {IC_95_superior:,.2f}]")
```
